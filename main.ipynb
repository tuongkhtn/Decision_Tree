```python
import os
import io
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo
```

```python
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)
wine_quality = fetch_ucirepo(id=186)
spambase = fetch_ucirepo(id=94) 
```

```python
dataset_shapes = {
    "Breast Cancer Wisconsin Diagnostic": (breast_cancer_wisconsin_diagnostic.data.features.shape[0], breast_cancer_wisconsin_diagnostic.data.features.shape[1]),
    "Wine Quality": (wine_quality.data.features.shape[0], wine_quality.data.features.shape[1]),
    "Spambase": (spambase.data.features.shape[0], spambase.data.features.shape[1])
}
```

```python
dataset_names = list(dataset_shapes.keys())
feature_counts = [shape[1] for shape in dataset_shapes.values()]

plt.figure(figsize=(8, 5))
plt.bar(dataset_names, feature_counts, color=['skyblue', 'lightgreen', 'salmon'])

plt.title("Số lượng đặc trưng của các dataset", fontsize=14)
plt.xlabel("Dataset", fontsize=12)
plt.ylabel("Số lượng đặc trưng", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig("feature_counts.png")

plt.show()
```

```python
dataset_names = list(dataset_shapes.keys())
sample_counts = [shape[0] for shape in dataset_shapes.values()]

plt.figure(figsize=(8, 5))
plt.bar(dataset_names, sample_counts, color=['skyblue', 'lightgreen', 'salmon'])

plt.title("Số lượng mẫu của các dataset", fontsize=14)
plt.xlabel("Dataset", fontsize=12)
plt.ylabel("Số lượng mẫu", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig("sample_counts.png")

plt.show()
```

```python
root_path = os.getcwd()

bcw_path = os.path.join(root_path, r"Breast_Cancer_Wisconsin\outputs\train_80_test_20\bcw_train_80_test_20_depth_accuracy.csv")
wq_path = os.path.join(root_path, r"Wine_Quality\outputs\train_80_test_20\wq_train_80_test_20_depth_accuracy.csv")
sb_path = os.path.join(root_path, r"Spambase\outputs\train_80_test_20\sb_train_80_test_20_depth_accuracy.csv")

bcw_df = pd.read_csv(bcw_path)
wq_df = pd.read_csv(wq_path)
sb_df = pd.read_csv(sb_path)

plt.figure(figsize=(10, 6))

plt.plot(bcw_df.columns[1:], bcw_df.iloc[0, 1:].values, marker='o', label="Breast Cancer", color="skyblue")
plt.plot(wq_df.columns[1:], wq_df.iloc[0, 1:].values, marker='o', label="Wine Quality", color="lightgreen")
plt.plot(sb_df.columns[1:], sb_df.iloc[0, 1:].values, marker='o', label="Spambase", color="salmon")

plt.title("Accuracy theo max_depth cho các dataset", fontsize=14)
plt.xlabel("max_depth", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig("accuracy_max_depth.png")

plt.show()
```
